```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
```

```python
models = [
    (
        'Logistic Regression',
        {'max_iter': 100, 'penalty': 'l2'},
        LogisticRegression(),
        (x_train, y_train),
        (x_test, y_test)
    ),
    (
        'KNN',
        {'n_neighbors': 11, 'leaf_size':30 },
        KNeighborsClassifier(),
        (x_train, y_train),
        (x_test, y_test)  
    ),
    (
        'Decision Tree Classifier',
        {'max_depth': 9, 'min_samples_leaf':727, 'min_samples_split': 2181 },
        DecisionTreeClassifier(),
        (x_train, y_train),
        (x_test, y_test)
    ),
    (
        'Gradient Boosting Classifier',
        {'n_estimators': 50, 'learning_rate': 0.1},
        GradientBoostingClassifier(),
        (x_train, y_train),
        (x_test, y_test)
    ),
    (
        'Random Forest Classifier',
        {'n_estimators': 501,'max_depth': None},
        RandomForestClassifier(),
        (x_train, y_train),
        (x_test, y_test)
    )
    
    
]
```

```python
reports = []

for model_name, params, model, train_set, test_set in models:
    x_train = train_set[0]
    y_train = train_set[1]
    x_test = test_set[0]
    y_test = test_set[1]
    model.set_params(**params)
    model.fit(x_train, y_train)
    score = model.score(x_test, y_test)
    
    reports.append((model_name, score))
```

```python
reports
```

```python
reports[1][1]
```

```python
mlflow.set_experiment('Classifier 2nd Experiment')
mlflow.set_tracking_uri('http://127.0.0.1:5000')
for i, element in enumerate(models):
    model_name = element[0]
    params = element[1]
    model = element[2]
    score = reports[i][1]
    
    with mlflow.start_run(run_name = model_name):
        mlflow.log_params(params)
        mlflow.log_metric('score',score)
        mlflow.sklearn.log_model(model, "model")
```
