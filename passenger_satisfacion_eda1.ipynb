```python
#Age Group
group_1 = np.arange(7, 13, 1)
group_2 = np.arange(13, 20, 1)
group_3 = np.arange(20, 36, 1)
group_4 = np.arange(36, 51, 1)
group_5= np.arange(51, 66, 1)
group_6 = np.arange(66, 86, 1)
```

```python
df.info()
```

```python
for i in range(df.shape[0]):
    if df['Age'].iloc[i] in group_1:
        df.loc[i, 'age_group']= 1
    elif df['Age'].iloc[i] in group_2:
        df.loc[i, 'age_group']= 2
    elif df['Age'].iloc[i] in group_3:
        df.loc[i, 'age_group'] = 3
    elif df['Age'].iloc[i] in group_4:
        df.loc[i, 'age_group'] = 4
    elif df['Age'].iloc[i] in group_5:
        df.loc[i, 'age_group'] = 5
    else:
        df.loc[i, 'age_group'] = 6
```

### Analyzing Ratings by Passengers

#### Rating of 'Boarding Services' by Male passengers

```python
plt.figure(figsize=(15, 4))
sns.lineplot(df[df['Gender']=='Male'].groupby('age_group')[boarding_service].mean())
```

```python
plt.figure(figsize=(15, 4))
sns.lineplot(df[df['Gender']=='Female'].groupby('age_group')[boarding_service].mean())
```
